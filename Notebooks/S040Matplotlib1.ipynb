```python
a = pd.Series(range(10)) *10
print(a)
b = pd.Series(range(10))**3

plt.plot(a.index, a.values, label = "line")
plt.plot(a.index, b.values, 'ro', label ="dot")

plt.xticks(a.index)
plt.title("First plot")
plt.xlabel("x-axis")
plt.ylabel("vertical axis")
plt.legend()
plt.show()
```

```python
a.plot(kind='line')
b.plot(kind='line')
plt.show()
```

```python
# plots each graph on different window
a.plot(kind='line')
#plt.clf() # clear figure
plt.show()

b.plot(kind='line')
plt.show()
```

```python
## plot multiple graphs on same figure
data =  pd.Series([1,4,3,5,6,5,4,1,2,3,4]) ** 2
data2  = data/2

plt.figure(figsize=(16,3))

plt.subplot(1,2, 2)
plt.plot(data.index,data.values)

plt.subplot(1,2, 1)
plt.plot(data2.index,data2.values, 'g^')
plt.axhline(10)
plt.text(2,5,"ab")
#plt.axvline(2)
plt.axvline(4, 0, 0.8)
plt.xticks(data2.index)
plt.savefig('graphs.png')
plt.show()
```

```python
x =       ["M","F"]
survived = [10,20]
dead =     [25,8]
plt.bar(x,survived, label="survived")
plt.bar(x,dead, bottom=survived, label="died")
plt.legend()
plt.show()
```

```python
x =       ["M","F"]
survived = [10,20]
dead =     [25,8]
plt.bar(x,survived, label="survived", alpha=0.5)
plt.bar(x,dead, label="died", alpha=0.5)
plt.legend()
plt.show()
```
