```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Assignments/gaurav/titanic_filtered.csv")
```

```python
print(df.head())
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.age.index, df.age.values)
plt.show()
```

```python
s = df.gender.value_counts()
print(s, type(s))

plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
plt.bar(s.index, s.values, width=0.9)

plt.subplot(2,2,2)
s.plot(kind='bar')


plt.subplot(2,2,3)
plt.barh(s.index, s.values)

plt.subplot(2,2,4)
s.plot(kind='barh')
plt.show()
```

```python
s = df.gender.value_counts()
print(s, type(s))

plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
plt.pie(s.values, labels=s.index, explode =[0.0,0.2])

plt.subplot(1,2,2)
s.plot(kind='pie')

plt.show()
```

```python
grp = df.groupby(by = ["survived", "gender"])
print(grp.count().pclass)
grp.count().pclass.plot(kind="bar")
plt.show()

plt.bar()
```

```python
grp = df.groupby(by = ["survived", "gender"])
print(grp.count().pclass)
grp.count().pclass.plot(kind="bar")
plt.show()

plt.bar()
```

```python
grp = df.groupby(by = ["survived", "gender"])
r = grp.count().pclass
print(r, type(r))

r.plot(kind="bar")
plt.show()
```

```python
grp = df.groupby(by = ["survived", "gender"])
r = grp.count().pclass
print(r)

print()
r1 = r.unstack()
print(r1)
r1.plot(kind="bar")
plt.show()
```

```python
grp = df.groupby(by = ["survived", "gender"])
r = grp.count().pclass
print(r)

print()
r1 = r.unstack()
print(r1)
r1.plot(kind="bar", stacked=True)
plt.show()
```

```python
df = pd.DataFrame({"A":[10,20], "B":[4,2], "C":[5,10]})
print(df)
df.plot(kind='bar')
plt.show()

df.plot(kind='bar', stacked=True)
plt.show()
```

```python
x=["M", "F",'C']
survived=pd.Series([10,20,30])
dead=pd.Series([25,15,5])
ill=pd.Series([5,10,15])

plt.figure(figsize=(11,5))
plt.bar(x,survived,label='survived')
plt.bar(x,dead,bottom=survived, label='dead')
plt.bar(x,ill,bottom=dead+survived, label='ill')
plt.legend()
```
