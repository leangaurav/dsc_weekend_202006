```python
# Read sheet "Australia by Residence" from Australia.xlsx
# Ignore the first 20 rows
import pandas as pd
import numpy as np
df=pd.read_excel("Australia.xlsx",sheet_name="Australia by Residence",skiprows=20)
print(df)

#There are 448 rows and 43 columns.
```

```python
# List all Column and Row indexes(Print their names)
print(df.columns)
print(df.columns.values)
print(df.index)
print(df.index.values)
```

```python
# Dataset Analysis
# Describe the dataset
print(df.describe(include='all'))
```

```python
# Use values and index attributes on both Row and Column Indexes
print(df.columns.values)
print(df.columns.get_loc('DevName'))
print(df.index.values)
```

```python
# Check Shape, Datatypes and other attributes and observe the datatype of each columns' data
print(df.shape)
print(df.dtypes)
print(df.RegName.dtypes)
```

```python
# Drop
# Delete all Data where the Type = "Emigrants" or Type is not "Immigrants" (Row Deletion)
#There are 448 rows and 43 columns.If you drop Emigrants (224), then only 224 rows will be left.
df1=df[df.Type!='Emigrants']
print(df1.head())
print()
print()

df=df[df.Type=='Immigrants']
print(df.head())
```

```python
# Find columns for which more than 50% data is missing and drop them (Column Deletion)
#After dropping Emigrants we have 224 rows left
print((df.isna().sum()).to_frame().T)
#There are few columns which has ".." values we would drop them as well but before that we will have to convert them into NAN
df.replace('..',np.nan, inplace=True)
print((df.isna().sum()).to_frame().T)
df.dropna(thresh=(224*0.5),axis=1,inplace=True)
print((df.isna().sum()).to_frame().T)
# df.dropna(axis=1,inplace=True)
# print()
# print(df)
```

```python
# Drop Unnessary Columns: Type, Coverage, Area, Reg, Dev (Column deletion) Your columns 
# should now look like "OdName,AreaName,RegName,DevName,1980 ... 2008"
df.drop(columns=['Type', 'Coverage', 'AREA', 'REG', 'DEV'], inplace=True)
print(df)
```

```python
# Rename columns :OdName -> Country; RegName -> Region
df.rename({'OdName': 'Country','RegName': 'Region'}, axis=1, inplace=True)
print(df.head())
print(df.describe(include='all'))
```
